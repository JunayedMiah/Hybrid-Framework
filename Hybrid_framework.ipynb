```python
## Import all relevant packages
import numpy as np
from sklearn.decomposition import KernelPCA
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt


## Generate synthetic nonlinear data
X, y = make_circles(n_samples = 1000, factor = 0.3, noise = 0.05, random_state= 0)

## Apply Kernel PCA
kpca = KernelPCA(n_components = 2, kernel = 'rbf' , gamma = 15)
X_kpca = kpca.fit_transform(X)

## Apply Spectral Clustering on the transformed data
n_clusters = 2
spectral = SpectralClustering(n_clusters = n_clusters, affinity = 'nearest_neighbors', n_neighbors = 10)
y_spectral = spectral.fit_predict(X_kpca)

## Plot the results
plt.figure(figsize= (14, 8))

#Original Data
plt.subplot(1, 2, 1)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.title('Original Data')

## Data after Kernel PCA and Spectral Clustering 
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_spectral, cmap='viridis')
plt.title('Kernel PCA and Spectral Clustering')


plt.show()
```
